1. Cargar y visualizar los primeros 5 registros del archivo 01-22-2020.csv.

```python
%pylab inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import requests
```

```python
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/refs/heads/master/csse_covid_19_data/csse_covid_19_daily_reports/"
```

```python
print(url)
```

```python
dframes = []
for i in range(1,32):
  if i < 10:
    dt = f"08-0{i}-2021.csv"
    dframes.append(dt)
  else:
    dt = f"08-{i}-2021.csv"
    dframes.append(dt)
print(dframes)
```

```python
urldf = [url+ dat for dat in dframes]
print(urldf)
```

```python
templist =[]
for a in urldf:
    try:
        tempdf = pd.read_csv(a)
        templist.append(tempdf)
    except Exception as e:
        print(f"Error al leer {a}: {e}")

```

```python
DFAgosto2021 = pd.concat(templist, ignore_index=True)
```

5. Eliminar columnas irrelevantes (por ejemplo, códigos FIPS o coordenadas si no se usarán).

```python
DFAgosto2021.drop(columns=["FIPS","Lat","Long_"],inplace=True)
```

6. Estandarizar nombres de columnas (usar formato snake_case).

```python
DFAgosto2021.columns = DFAgosto2021.columns.str.lower().str.replace(" ","_").str.replace("-","_")
```

7. Homogeneizar nombres de países (ej. “US” → “United States”)

```python
DFAgosto2021.loc[DFAgosto2021["country_region"] == "US", "country_region"] = "United States"
```

8. Convertir la columna Last_Update al formato YYYY-MM-DD.

```python
DFAgosto2021["last_update"] = DFAgosto2021["last_update"].dt.date
```

9. Crear una columna active_cases = Confirmed - Deaths - Recovered

```python
DFAgosto2021["active_cases"] = DFAgosto2021["confirmed"] - DFAgosto2021["deaths"] - DFAgosto2021["recovered"]
```
